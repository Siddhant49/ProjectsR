# Item-Based Collaborative Filtering

```python
import pandas as pd
import numpy as np 

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('D:\DataScience-Python3\ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('D:\DataScience-Python3\ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)

ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.tail()
```

```python
newUR = np.asarray(userRatings).astype(np.float64)
newUR.shape
```

```python
#outliers=[]
for row in range(944):
    count=0
    for col in range(1644):
        if (np.isnan(newUR[row][col]) == False):
            count+=1
        if (count==100):
            userRatings.drop(index=row)
userRatings.shape
            
```

```python
corrMatrix = userRatings.corr(method='kendall', min_periods=150)
corrMatrix.dropna()

corrMatrix.head()
```

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

```python
simCandidates = pd.Series(dtype=object)

for i in range(0, len(myRatings.index)):
    print ("Adding sims for " + myRatings.index[i] + "...")
   
    sims = corrMatrix[myRatings.index[i]].dropna()
    
    sims = sims.map(lambda x: x * myRatings[i])
    
    simCandidates = simCandidates.append(sims)
    
print ("sorting...")

simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()

simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
